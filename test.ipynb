```python
import pandas as pd
from sklearn import ensemble
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('data/train.csv')
X_train, y_train = df.iloc[:,:-1], df.iloc[:,-1]
print(y_train[y_train==1].count())
print(y_train[y_train==0].count())
X_train
```

```python
cat_cols = ['person_gender', 'person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file']

ohe = OneHotEncoder(drop='if_binary')
classifier = ensemble.RandomForestClassifier(n_estimators=200, criterion='entropy')

transformer = ColumnTransformer(
    [('cat_trans', ohe, cat_cols)], 
    remainder='passthrough')
transformer.fit(X_train)

X_train_transformed = transformer.transform(X_train)
```

```python
classifier.fit(X_train_transformed, y_train)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
y_train_pred = classifier.predict(X_train_transformed)
print(f'Train Accuracy: {accuracy_score(y_train, y_train_pred)}')
print(f'Train F1: {f1_score(y_train, y_train_pred)}')
print(f'Train Precision: {precision_score(y_train, y_train_pred)}')
print(f'Train Recall: {recall_score(y_train, y_train_pred)}')
```

```python
df = pd.read_csv('data/test.csv')

X_test, y_test = transformer.transform(df.iloc[:,:-1]), df.iloc[:,-1]

y_test_pred = classifier.predict(X_test)

print(f'Train Accuracy: {accuracy_score(y_test, y_test_pred)}')
print(f'Test F1: {f1_score(y_test, y_test_pred)}')
print(f'Test Precision: {precision_score(y_test, y_test_pred)}')
print(f'Test Recall: {recall_score(y_test, y_test_pred)}')
```
